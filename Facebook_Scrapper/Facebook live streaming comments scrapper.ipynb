```python
from selenium import webdriver
import time
import pandas as pd
from selenium.webdriver.common.keys import Keys
links=[]
comment=[]
name1=[]
post_url=[]
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.facebook.com/pg/hillreporter/posts/')
driver.find_element_by_name('email').send_keys("amit1.indybytes@gmail.com")
driver.find_element_by_name('pass').send_keys("amit@1234")
driver.find_element_by_xpath("//input[@value='Log In']").click()
while True:
    last_height = driver.execute_script("return document.body.scrollHeight")
    links=driver.find_elements_by_xpath('//a[@class="_3hg- _42ft"]')
    for i in links:
        try:
            lik=i.get_attribute('href')
#             print(lik)
    #             print(d)
            if lik not in links:
                links.append(lik)
        except:
            pass
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(4)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
for t in links:
    try:
        driver.get(t)
        time.sleep(3)
        driver.execute_script("window.scrollTo(0, 800);")
        driver.find_element_by_xpath('//a[@class="_2pm3 _21q1 _p"]').click()
        driver.find_elements_by_xpath('//div[@class="_1ojv"]')[2].click()
        time.sleep(2)
        new=driver.find_elements_by_xpath('//span[@class=" _4ssp"]')[0]
        while True:
            try:
                new.click()
            except:
                break
        comments=driver.find_elements_by_xpath('//span[@class="_3l3x"]/span')
        name=driver.find_elements_by_xpath('//div[@class="_72vr"]/a')
        for i,j in zip(comments,name):
#             print(i.text)
#             print(j.text)
            comment.append(i.text)
            name1.append(j.text)
            post_url.append(t)
    except:
        pass
if len(post_url)>len(comment):
    big=len(post_url)
else:
    big=len(comment)
if big>len(name1):
    big=big
else:
    big=len(name1)
print(big)
for i in range(big):
    if i>=len(post_url):
        
        post_url.append("None")
    if i>=len(comment):
        comment.append("None")
    if i>=len(name1):
        name1.append("None")
df = pd.DataFrame(
    {'Post_url': post_url,'Comment':comment,"Name":name1})
print(df)
df.to_csv("All_Comments.csv", mode='a',encoding='utf-32',index=False,header=True,sep='\t', escapechar='\\')
```
